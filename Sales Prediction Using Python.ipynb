## Splitting the data into dependent and independent Variables

## Feature Selection

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
x_pca=pca.fit_transform(x)
```

## Splitting the data in to train and test 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=0)
```

# Sales predicition using Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mse
```

```python
r2
```

# Sales predicition using SVR with Hyperparameter tuning

```python
from sklearn.svm import SVR
```

```python
# Define a range of C values to test
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}
```

```python
# Define a custom scoring metric (e.g., mean squared error)
from sklearn.metrics import mean_squared_error, make_scorer
svr = SVR(kernel='linear')
scoring = make_scorer(mean_squared_error, greater_is_better=False)
```

```python
from sklearn.model_selection import GridSearchCV
# Perform grid search with cross-validation

grid_search = GridSearchCV(svr, param_grid, cv=5, scoring=scoring)
grid_search.fit(x_train, y_train)

# Get the best C value
best_C = grid_search.best_params_['C']
```

```python
svr = SVR(kernel='linear', C=best_C)
```

```python
svr.fit(x_train, y_train)
```

```python
y_pred =svr.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mse
```

```python
r2
```

# Sales prediction using Randomforest Regressor

# Sales prediction using Randomforest Regressor with Hyperparameter tuning

```python
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Create the Random Forest Regressor model
rf_regressor = RandomForestRegressor(random_state=0)

# Define a custom scoring metric (e.g., mean squared error)
scoring = make_scorer(mean_squared_error, greater_is_better=False)


```

```python
# Perform grid search with cross-validation
grid_search = GridSearchCV(rf_regressor, param_grid, cv=5, scoring=scoring)
```

```python
#searching for best parameters for the model
grid_search.fit(x_train, y_train)
```

```python
best_params = grid_search.best_params_
```

```python
best_rf_regressor = RandomForestRegressor(random_state=0, **best_params)
best_rf_regressor.fit(x_train, y_train)

# Make predictions using the best model
y_pred = best_rf_regressor.predict(x_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(mse)
print(r2)
```

Observation:
Random forest regressor give more accurate results than other algorithms.
